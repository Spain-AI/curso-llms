## Load data

```python
import yfinance as yf

stocks = ['AAPL', 'MSFT', 'AMZN', 'GOOGL', 'GOOG', 'META', 'TSLA']
df_base = yf.download(stocks, start='2010-01-01', end='2020-12-31')
```

```python
df_base
```

## Calculate metrics

```python
df = df_base['Close']

df = (df
 .loc['2010':]
 .groupby(df.index.year).pct_change(fill_method=None).add(1)
 .groupby(df.index.year).cumprod().sub(1)
 .resample('1YE').last()
 .mul(100)
)
```

```python
df.style
```
