```python
import chromadb
from chromadb.utils import embedding_functions

CHROMA_DATA_PATH = "../chromadb/full"
EMBED_MODEL = "all-MiniLM-L6-v2"
COLLECTION_NAME = "articles"

client = chromadb.PersistentClient(path=CHROMA_DATA_PATH)
```

```python
embedding_func = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name=EMBED_MODEL
)

collection = client.create_collection(
    name=COLLECTION_NAME,
    embedding_function=embedding_func,
    metadata={"hnsw:space": "cosine"},
    get_or_create=True
)
```

```python
documents
```

```python
texts = [doc['summary'] for doc in documents]
ids = [doc['url'] for doc in documents]
metadatas = [{
    "title": doc['title'],
    "url": doc['url'],
    "time_published": doc['time_published'],
    "authors": doc['authors'],
    "summary": doc['summary'],
    "banner_image": doc['banner_image'],
    "source": doc['source'],
    "category_within_source": doc['category_within_source'],
    "source_domain": doc['source_domain'],
    "topics": doc['topics'],
    "overall_sentiment_score": doc['overall_sentiment_score'],
    "overall_sentiment_label": doc['overall_sentiment_label'],
    "ticker_sentiment": doc['ticker_sentiment'],
    "pct_change_value": pct_change_value,
    'date': str(date)
} for doc in documents]
```

```python
l = []
for i in metadatas:
    data = {}
    for key, value in i.items():
        # check if value is any of str, int, float or bool
        if not isinstance(value, (str, int, float, bool)):
            data[key] = json.dumps(value)
        else:
            data[key] = value
    l.append(data)
```

```python
collection.add(
    documents=texts,
    ids=ids,
    metadatas=l
)
```

```python
results = collection.query(
    query_texts=['Apple stock up because of release'],
    n_results=1
)

results
```

```python
results = collection.query(
    query_texts=['Apple stock up because of release'],
    n_results=2,
    include=['documents', 'distances']
)

results
```
